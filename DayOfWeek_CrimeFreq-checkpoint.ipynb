```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
crime=pd.read_csv('samplecrime.csv')
crime
```

```python
type(crime)
```

```python
crime['DayOfWeek']
crime1=crime[['Category','DayOfWeek','Date','Time']]
crime1
```

```python
count_dayofweek=crime['DayOfWeek'].value_counts()
count_dayofweek
```

```python
type(count_dayofweek)
```

```python
df=pd.DataFrame(count_dayofweek)
df
```

```python
# reset
DF1=df.reset_index()
DF1
```

```python
# rename
DF1.columns=['DayOfWeek','Freq of crime']
DF1
Sample=DF1.head(7)
Sample
```

```python
# create bar chart
plt.bar(Sample['DayOfWeek'],Sample['Freq of crime'],color="Red")
plt.xlabel("Day of Week")
plt.ylabel("Frequency of crime")
plt.title("The frequencies of crime each day of the week")
plt.show()
```
